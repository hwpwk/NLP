# ファイルを横に連結

#### 1.まずは「氏名」カラムのみ抽出したマスタテーブルを用意

#### 2.for文で読み込むcsvファイルの「売上(万円)」カラムのみ抽出して、氏名をキーにしながら横に結合していく

#### 3.この際、
#### 【最初に結合するのはcsvファイルとマスタテーブル】
#### 【それ以降はマスタテーブルと結合したデータフレーム】に連続で結合していくため条件分岐が必要

```python
import pandas as pd

import re
```

### マスタテーブルの用意

```python
df_master = pd.read_csv('Sales_Ranking_201706.csv', usecols=['氏名'])
df_master
```

### 「glob」を使ってフォルダ内すべてのcsvファイルリストを取得

```python
import glob
files = glob.glob('*.csv')
```

### マスタテーブルの「氏名」をキーにしてcsvファイルを１つずつ横に連結していく
この際、

【最初に結合するのはcsvファイルとマスタテーブル】

【それ以降はマスタテーブルと結合したデータフレーム】に連続で結合していくため条件分岐が必要

```python
for file in files:
    df_sales = pd.read_csv(file, usecols=['氏名', '売上(万円)'])
    number = re.sub('\D', '', file)#「Sales_Ranking_201706.csv」なら「201706」を抽出する
    df_sales = df_sales.rename(columns={'売上(万円)':'売上(万円)_' + str(number)})
    if file == files[0]:#if file == 'Sales_Ranking_201706.csv'と同じ意味
        df_merge = pd.merge(df_master, df_sales, on='氏名', how='outer')
    else:
        df_merge = pd.merge(df_merge, df_sales, on='氏名', how='outer')

df_merge
```
