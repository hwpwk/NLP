```python
adopt_word_tfidf_merge_df = pd.read_pickle('tfidf_df.pkl')
adopt_word_tfidf_merge_df.head()
```

#### 全テキスト内容×全単語のデータフレーム を作成(TFIDF値採用)

```python
document_word_tfidf_df = adopt_word_tfidf_merge_df[['内容', 'original_type', 'tf_idf']]
document_word_tfidf_df.head()
```

```python
model_tfidf_df = pd.pivot_table(document_word_tfidf_df, index='内容', columns='original_type', values='tf_idf')
# nanは0に変換する
model_tfidf_df = model_tfidf_df.fillna(0)
model_tfidf_df# 完成
```

## 品詞で特徴量を絞る
今回は記号と助詞の単語以外を特徴量とする

```python
#  TFIDFの標準偏差の値を格納したファイルの読み込み
std_df = pd.read_pickle('std_df.pkl')
std_df.head()
```

### 正規表現を使って特定の品詞「以外」の単語のみ抽出
下記サイトを使い正規表現のパターンマッチ表現を探索する

https://regex101.com

正規表現一覧は下記

http://gimite.net/help/devas-ja/all_regex.html

```python
# 正規表現によって文字列を抽出する場合はstr.match関数を使用
# 今回は記号＆助詞「以外」の単語抽出なので「~」チルダをつける
# 比較演算子を使うときは条件ごとに括弧で囲む（括弧がないとエラー） よって、チルダは括弧の前につける
# 参考URL:https://note.nkmk.me/python-pandas-multiple-conditions/
except_symbol_df = std_df[~(std_df['original_type'].str.match('.*_記号-.*')) & ~(std_df['original_type'].str.match('.*_助詞-.*'))]
except_symbol_df
```

```python
# TFIDFの標準偏差0.001以上の単語を抽出
except_symbol0001_df = except_symbol_df[except_symbol_df['std_tf_idf'] >= 0.001]
except_symbol0001_df.head()
```

```python
# 抽出した単語部分のカラム１つのみ指定してSeries型にした上でそれをリスト型にする
feature_list = except_symbol0001_df['original_type'].tolist()
feature_list
```

```python
# 単語が格納されたリストを全文書×全単語データフレーム(model_tfidf_df)のカラムに指定するとカラムがリストに格納されていた単語に変換される
except_symbol0001_model_df = model_tfidf_df[feature_list]
except_symbol0001_model_df# 完成
```

```python
# データフレームをpklファイルとして保存
except_symbol0001_model_df.to_pickle('except_symbol0001_model_df.pkl')
```
