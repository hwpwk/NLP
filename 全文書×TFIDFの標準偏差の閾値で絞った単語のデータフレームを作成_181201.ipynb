```python
adopt_word_tfidf_merge_df = pd.read_pickle('tfidf_df.pkl')
adopt_word_tfidf_merge_df.head()
```

#### 全テキスト内容×全単語のデータフレーム を作成(TFIDF値採用)

```python
document_word_tfidf_df = adopt_word_tfidf_merge_df[['内容', 'original_type', 'tf_idf']]
document_word_tfidf_df.head()
```

```python
model_tfidf_df = pd.pivot_table(document_word_tfidf_df, index='内容', columns='original_type', values='tf_idf')
# nanは0に変換する
model_tfidf_df = model_tfidf_df.fillna(0)
model_tfidf_df# 完成
```

## TFIDFの標準偏差の値で特徴量を絞る

```python
#  TFIDFの標準偏差の値を格納したファイルの読み込み
std_df = pd.read_pickle('std_df.pkl')
std_df.head()
```

```python
# TFIDFの標準偏差0.001以上の単語を抽出
std0001_df = std_df[std_df['std_tf_idf'] >= 0.001]
std0001_df.head()# ６８単語中27単語抽出される
```

```python
# 抽出した単語部分のカラム１つのみ指定してSeries型にした上でそれをリスト型にする
feature_list = std0001_df['original_type'].tolist()
feature_list
```

```python
# 単語が格納されたリストを全文書×全単語データフレーム(model_tfidf_df)のカラムに指定するとカラムがリストに格納されていた単語に変換される
std001_model_tfidf_df = model_tfidf_df[feature_list]
std001_model_tfidf_df# 完成
```
